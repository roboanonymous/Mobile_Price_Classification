```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
import seaborn as sns
```

```python
data = pd.read_csv("train.csv")
```

```python
data
```

```python
x = data.iloc[:, 0:20]
y = data.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
best_features = SelectKBest(score_func = chi2, k=10)
```

```python
fit = best_features.fit(x,y)
```

```python
fit.scores_
```

```python
dfscores= pd.DataFrame(fit.scores_)
dfcolumns =  pd.DataFrame(x.columns)
```

```python
featurescores = pd.concat([dfcolumns, dfscores], axis=1)
```

```python
featurescores.columns = ['Features', 'score']
```

```python
feature_ten = featurescores.sort_values(by = 'score', ascending = False).head(10)['Features'].values
```

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
```

```python
model = RandomForestClassifier()
model.fit(x,y)
```

```python
model.feature_importances_
```

```python
featureImportance = pd.DataFrame(model.feature_importances_, index= x.columns, columns = ['Importance'])
```

```python
featureImportance = featureImportance.sort_values(by = 'Importance' , ascending = False)
```

```python
plt.figure(figsize= (20,10))
plt.bar(featureImportance.index, featureImportance['Importance'])
plt.show()
```
